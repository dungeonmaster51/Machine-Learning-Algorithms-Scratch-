```python
split = int(0.85*X.shape[0])
X_train = X[:split]
y_train = y[:split]

X_test = X[split:]
y_test = y[split:]
```

```python
# hypothesis function
def hypothesis(x, w):
    return x*w[1] + w[0]

# error compute
def get_error(y_true, x, w,a):
    err = 0.0
    
    # Get number of examples
    m = x.shape[0]
    for ix in range(m):
        # for each example compute the square error
        err += (y_true[ix] - hypothesis(x[ix], w))**2 + a*abs(w)
    # Get mean sq. error
    err = err / (2*m)
    return err

# error gradient
def get_grads(y_true, x, w,a):
    grad_00 = 0.0
    grad_01 = 0.0
    
    # Get number of examples
    m = x.shape[0]
    
    for ix in range(m):
        # for each example compute the gradients
        a1 = a
        a2 = a
        if w[0] < 0:
            a1 = -a
        if w[1] < 0:
            a2 = -a
        grad_00 += (y_true[ix] - hypothesis(x[ix], w))*(-1) + a1
        grad_01 += (y_true[ix] - hypothesis(x[ix], w))*(-1*x[ix]) + a2  
    
    # Get mean grads
    grad_00 = grad_00 / m
    grad_01 = grad_01 / m
    return [grad_00, grad_01]

# update weights using grad. desc.
def grad_descent(x, y_true, w, learning_rate=0.1,a = .01):
    error = get_error(y_true, x, w,a)
    [grad_00, grad_01] = get_grads(y_true, x, w,a)
    
    w[0] = w[0] - learning_rate*grad_00
    w[1] = w[1] - learning_rate*grad_01
    
    return error, w
```

```python
W = np.array([0.1, 0.1])
```

```python
import numpy as np
import matplotlib.pyplot as plt
loss_ridge = []
loss_modulus = []
loss = []
```

```python
for ix in range(100):

    plt.scatter(X[:, 0], y)

    y_0 = hypothesis(-3, W)
    y_1 = hypothesis(3, W)
    
    

    l, W = grad_descent(X_train, y_train, W, learning_rate=0.01)
    loss_ridge.append(l)
plt.plot([-3, 3], [y_0, y_1], 'r*-')
plt.show()
# print loss

print len(loss_ridge)
```
